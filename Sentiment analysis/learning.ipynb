```python
# Simple positive and negative word lists
positive_words = ['win', 'profit', 'good', 'poa', 'success', 'safi', 'hela', 'faida']
negative_words = ['bad', 'mbaya', 'lose', 'lost', 'zero', 'fail', 'hauna']

# Function to score sentiment
def get_sentiment(text):
    words = text.lower().split()
    pos = sum(word in positive_words for word in words)
    neg = sum(word in negative_words for word in words)

    if pos > neg:
        return "Positive"
    elif neg > pos:
        return "Negative"
    else:
        return "Neutral"

# Analyze each tweet and print the result
for tweet in cleaned_tweets:
    sentiment = get_sentiment(tweet)
    print(f"{sentiment}: {tweet}")
```

```python
# Count the sentiment labels
sentiment_counts = {
    "Positive": 0,
    "Negative": 0,
    "Neutral": 0
}

# Count each sentiment
for tweet in cleaned_tweets:
    sentiment = get_sentiment(tweet)
    sentiment_counts[sentiment] += 1
```

```python
import matplotlib.pyplot as plt

# Bar chart
labels = sentiment_counts.keys()
values = sentiment_counts.values()

plt.figure(figsize=(6, 4))
plt.bar(labels, values, color=['green', 'red', 'gray'])
plt.title("Sentiment Analysis of Tweets")
plt.xlabel("Sentiment")
plt.ylabel("Number of Tweets")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
